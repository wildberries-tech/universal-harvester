Поскольку шифрование происходит на встроенных в приложение функциях, необходимо сформировать болванку current_state.

```python
# dummy current state

ENGINE_PATH = ""
STORAGE_PATH = ""
ITSELF_LINK = ""

DB_PATH = ""
STORAGE_PATH = ""
APP_NAME = "Universal Harvester object creator"
APP_VERSION = "1.0.0"

DUMMY_SESSION_ID = "00000000-0000-0000-0000-000000000000"
DUMMY_IP = "127.0.0.1"
DUMMY_PORT = 0
DUMMY_USERNAME = "dummy"

current_state = {
    "db_path":DB_PATH,
    "app_name":APP_NAME,
    "app_version":APP_VERSION,
    "main_session_id":DUMMY_SESSION_ID,
    "user_session_id":DUMMY_SESSION_ID,
    "client_ip_address":DUMMY_IP,
    "client_port":DUMMY_PORT,
    "username":"system",
    "master_key": MASTER_KEY,
    "codemirror_theme":'monokai',
    "storage_path":STORAGE_PATH,
    "engine_path":ENGINE_PATH,
    "itself_link":ITSELF_LINK
}
```

Необходимо заполнить dict конфигурации подключения программы к БД.

type -- отвечает за то, какая БД будет использоваться. На данный момент доступны 2 варианта: postgresql или sqlite3.

Для sqlite3 необходимо указать путь к файлу БД, для работы в контейнере логично будет расположить файл БД в подмонтируемом STORAGE.

Для postgresql необходимы параметры подключения.

```python


db_conf = {
    "type":"postgresql", # or sqlite3
    "sqlite3":{
        "db_path":"../app.db"
    },
    "postgresql":{
        "host":"1.2.3.4",
        "port":5432,
        "db_name":"",
        "login":"",
        "password":""
    }
}

encrypt_string = encrypt(base64.b64encode(json.dumps(db_conf).encode()).decode(), current_state)[3]
print(encrypt_string)
```

Подготовленный encrypt_string необходимо использовать в параметрах запуска.

Далее проверяем, что всё нормально расшифровывается.

```python
base64.b64decode(decrypt(encrypt_string, current_state)[3])
```
